```python
import wordcloud
from matplotlib import pyplot as plt
from IPython.display import display

punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "in", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just"]

def word_frequencies(filename):
    word_cloud = {}
    file = open(filename, 'rt')
    file_content = file.read()
    file_content = file_content.split()

    for word in file_content:
        word = word.lower()
        new_word = ''
        for char in word:
            if char.isalpha():
                new_word += char
        if new_word in uninteresting_words:
            continue
        if new_word not in word_cloud:
            word_cloud[new_word] = 0
        word_cloud[new_word] += 1
        
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(word_cloud)
    cloud.to_file("wordcloud.jpg")
    return cloud.to_array()
```

```python
filename = 'Word.txt'
myimage = word_frequencies(filename)
plt.figure(figsize=(15,7))
plt.imshow(myimage, interpolation='nearest')
plt.axis('off')
plt.show()
```
